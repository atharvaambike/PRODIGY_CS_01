```python
import tkinter as tk
from tkinter import messagebox

def caesar_cipher_encrypt(text, shift):
    encrypted_text = ""
    for char in text:
        if char.isalpha():
            shift_amount = shift % 26
            if char.islower():
                encrypted_text += chr((ord(char) - ord('a') + shift_amount) % 26 + ord('a'))
            else:
                encrypted_text += chr((ord(char) - ord('A') + shift_amount) % 26 + ord('A'))
        else:
            encrypted_text += char
    return encrypted_text

def caesar_cipher_decrypt(text, shift):
    decrypted_text = ""
    for char in text:
        if char.isalpha():
            shift_amount = shift % 26
            if char.islower():
                decrypted_text += chr((ord(char) - ord('a') - shift_amount) % 26 + ord('a'))
            else:
                decrypted_text += chr((ord(char) - ord('A') - shift_amount) % 26 + ord('A'))
        else:
            decrypted_text += char
    return decrypted_text

def perform_action():
    action = action_var.get()
    text = text_entry.get("1.0", tk.END).strip()
    try:
        shift = int(shift_entry.get().strip())
    except ValueError:
        messagebox.showwarning("Warning", "Shift key must be an integer.")
        return

    if action == 'ee':
        result = caesar_cipher_encrypt(text, shift)
        messagebox.showinfo("Result", f"Encrypted: {result}")
    elif action == 'dd':
        result = caesar_cipher_decrypt(text, shift)
        messagebox.showinfo("Result", f"Decrypted: {result}")
    else:
        messagebox.showwarning("Warning", "You've entered wrong. Please try again.")

# Initializing the main window
root = tk.Tk()
root.title("Caesar Cipher")

# Creating and placing widgets
action_var = tk.StringVar()

label = tk.Label(root, text="What do you want to perform?")
label.pack()

encrypt_radio = tk.Radiobutton(root, text="Encryption", variable=action_var, value='ee')
encrypt_radio.pack()

decrypt_radio = tk.Radiobutton(root, text="Decryption", variable=action_var, value='dd')
decrypt_radio.pack()

shift_label = tk.Label(root, text="Enter shift key:")
shift_label.pack()

shift_entry = tk.Entry(root)
shift_entry.pack()

text_label = tk.Label(root, text="Enter text:")
text_label.pack()

text_entry = tk.Text(root, height=5, width=40)
text_entry.pack()

perform_button = tk.Button(root, text="Perform Action", command=perform_action)
perform_button.pack()

exit_button = tk.Button(root, text="Exit", command=root.quit)
exit_button.pack()

# Main event loop
root.mainloop()
```
